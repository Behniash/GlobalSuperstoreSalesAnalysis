# Executive Summary

This project analyzes the Global Superstore dataset to identify the main drivers of sales and profitability.

The analysis shows that Technology is the most profitable category, while Furniture generates high sales but relatively low profit. High discounts are strongly associated with negative profitability, and approximately one-quarter of orders result in a loss.

Based on these findings, several business recommendations and potential A/B tests are proposed.

# Business Insights

## Insight 1
Technology generated the highest sales and the highest profit.

## Insight 2
Furniture has strong sales but relatively low profitability.

## Insight 3
24.5% of orders were loss-making.

## Insight 4
Medium and High discounts resulted in negative average profits.

## Insight 5
Q4 produced the highest sales and profit.

## Insight 6
Phones and Copiers are the strongest product categories.

## Insight 7
Turkey and Nigeria generated significant losses.

## Insight 8
Approximately 42.5% of customers generated 80% of total sales.

# Business Recommendations

## Recommendation 1

Review the pricing strategy for Furniture products.



## Recommendation 2

Reduce discounts above 25%.



## Recommendation 3

Increase marketing investment for Technology products.



## Recommendation 4

Investigate loss-making countries such as Turkey and Nigeria.



## Recommendation 5

Increase inventory before Q4.

# Conclusion

Technology represents the strongest growth opportunity for the company, while Furniture and high-discount transactions require strategic review.

Future efforts should focus on optimizing discount policies, improving profitability in loss-making regions, and validating recommendations through controlled A/B experiments.
